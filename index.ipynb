```python
import matplotlib.pyplot as plt

q25, q75 = np.percentile(dummy_df['price'], [25, 75])
iqr = q75 - q25

lower_bound = 5
upper_bound = 60

outliers = dummy_df['price'][(dummy_df['price'] < lower_bound) | (dummy_df['price'] > upper_bound)]

# Plot the box plot with outliers
plt.figure(figsize=(8, 6))
plt.boxplot(dummy_df['price'])
plt.scatter(np.ones_like(outliers), outliers, color='red', marker='o', label='Outliers')
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box Plot with Outliers')
plt.legend()
plt.show()




```
